# Important libraries to be used

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read the dataset1 and display visualization

```python
data1 = pd.read_csv('dataset1.csv')
data1.head()
```

## We use dropna function for clean the data

```python
data1.dropna(inplace=True)
data1
```

```python
temp_over_time = data1['dteday'].value_counts().sort_index()
temp1 = data1['temp'].value_counts().sort_index()
plt.figure(figsize=(10, 8))
plt.scatter(temp1.index,temp1.values,color='red')
plt.title('Monthly tempreature in 2023')
plt.xlabel('date')
plt.ylabel('temp')
plt.show()
```

# Read second datasets 

```python
data2 = pd.read_csv('dataset2.csv')
data2
```

## Drop unnecesary colomn Unnamed: 0

```python
data2.drop(columns=["Unnamed: 0"],axis=0,inplace=True)
print(data2)
```

```python
data2.dropna(inplace=True)
data2
```

## Merge Data1 and Two using merge method

```python
merged_data = pd.merge(data1,data2,on = "instant",how = "right")
print(merged_data.to_string())
```

# Read datasets3 

```python
data3 = pd.read_csv('dataset3.csv')
data3
```

```python
data3.dropna(inplace=True)
data3
```

## Using concate method we are merging two data

```python
main_dataset = pd.concat([merged_data,data3],ignore_index=True)
main_dataset
```

## download the main dataset

```python
main_file = main_dataset.to_csv("Main_file.csv",index=False)
```
